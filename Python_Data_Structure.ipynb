1. What are data structures and why are they important?
   - Think of data structures like containers—you use different ones based on what you’re trying to store and how you want to access it. They help keep data organized so your code can run faster and smoother.

2. Mutable vs Immutable Data Types:
   - Mutable stuff you can change, like lists—add, remove, shuffle things around. Immutable ones like strings or tuples? Once you make them, they stay that way. It’s like writing with a pen vs a pencil.

3. Difference between lists and tuples in Python:  
   - Lists are flexible—you can edit them, add stuff, remove stuff. Tuples are more like locked boxes. Once you create them, you can’t mess with them. Good for data you don’t want changing.

4. How do dictionaries store data?
   - Dictionaries use key-value pairs. It’s kinda like a phonebook where names (keys) give you numbers (values). Python uses a behind-the-scenes thing called a hash table to make lookups super quick.

5. Why use a set instead of a list?
   - If you don’t want duplicates and care more about checking if something exists, sets are perfect. Lists let you have repeated stuff, but sets are all about uniqueness.

6. What’s a string and how’s it different from a list?
   - A string is just a bunch of characters strung together—like a word or sentence. Lists can hold anything (numbers, words, even other lists). Also, strings can’t be changed once made; lists can.

7. How do tuples help keep data safe?
   - Since you can’t change a tuple after creating it, it’s a good way to protect data from accidental changes. It’s like saying, “Hands off—this data stays as it is.”

8. What’s a hash table and how do dictionaries use it?  
   - A hash table is a clever way to store and find data quickly. Python’s dictionaries use them so when you look something up by key, it doesn’t have to search the whole thing.

9. Can a list have different data types?  
   - Totally. You can mix numbers, strings, even other lists or dictionaries all in one list. Python’s pretty chill about that.

10. Why are strings immutable?
    - It’s mostly about safety and performance. When strings don’t change, Python can reuse them behind the scenes and avoid bugs from accidental edits.

11. Why choose a dictionary over a list sometimes?  
    - If you care more about what something is (like a key) rather than where it is (like a position), dictionaries are the way to go. They make data easier to find.

12. When is a tuple better than a list?  
    - Let’s say you have coordinates (x, y) or config settings that shouldn’t change—you’d go for a tuple. It’s about keeping things locked down.

13. How do sets deal with duplicates?  
    - They just don’t allow them. If you try to add the same thing again, it silently ignores it. Handy for keeping your data clean.

14. How does “in” work for lists vs dictionaries?  
    - In a list, “in” checks if a value exists. In a dictionary, it checks if a key exists. Easy mix-up!

15. Can you change a tuple’s items?  
    - Nope. Once it’s created, that’s it. You’d need to make a whole new one if you want something different.

16. What’s a nested dictionary?  
    - It’s like a dictionary inside another one. Think of a student’s profile that has personal info, and under that, a sub-dictionary for marks or subjects.

17. What’s the time complexity of accessing elements in a dictionary?  
    - Usually super fast—O(1). It’s like asking for a specific item and getting it instantly without searching around.

18. When would you pick a list over a dictionary?
    - If you only need to store values and the order matters, a list is simpler and makes more sense.

19. Why are dictionaries considered unordered?  
    - Earlier versions of Python didn’t keep track of order. Even now, you shouldn't rely on order unless you're on a recent version where insertion order is preserved.

20. List vs Dictionary – How do you find data?  
    - With a list, you go by position (like `list[0]`). With a dictionary, you ask for a specific key (like `dict["name"]`). More direct and descriptive.



```python
# 1. Print your name
print("Shivam")

# 2. Length of "Hello World"
print(len("Hello World"))

# 3. Slice first 3 characters
print("Python Programming"[:3])

# 4. Convert to uppercase
print("hello".upper())

# 5. Replace word
print("I like apple".replace("apple", "orange"))

# 6. List 1 to 5
print([1, 2, 3, 4, 5])

# 7. Append 10
lst = [1, 2, 3, 4]
lst.append(10)
print(lst)

# 8. Remove 3
lst = [1, 2, 3, 4, 5]
lst.remove(3)
print(lst)

# 9. Access 2nd element
print(['a', 'b', 'c', 'd'][1])

# 10. Reverse list
print([10, 20, 30, 40, 50][::-1])

# 11. Create tuple
print((100, 200, 300))

# 12. Second-to-last element
print(('red', 'green', 'blue', 'yellow')[-2])

# 13. Minimum in tuple
print(min((10, 20, 5, 15)))

# 14. Index of "cat"
print(('dog', 'cat', 'rabbit').index("cat"))

# 15. Check "kiwi"
print("kiwi" in ("apple", "banana", "cherry"))

# 16. Create set
print({'a', 'b', 'c'})

# 17. Clear set
s = {1, 2, 3, 4, 5}
s.clear()
print(s)

# 18. Remove 4
s = {1, 2, 3, 4}
s.remove(4)
print(s)

# 19. Union
print({1, 2, 3}.union({3, 4, 5}))

# 20. Intersection
print({1, 2, 3}.intersection({2, 3, 4}))

# 21. Dictionary with keys
print({"name": "John", "age": 25, "city": "New York"})

# 22. Add "country"
d = {'name': 'John', 'age': 25}
d['country'] = "USA"
print(d)

# 23. Access "name"
print({'name': 'Alice', 'age': 30}['name'])

# 24. Remove "age"
d = {'name': 'Bob', 'age': 22, 'city': 'New York'}
del d['age']
print(d)

# 25. Check "city"
print("city" in {'name': 'Alice', 'city': 'Paris'})

# 26. Create list, tuple, dict
print([1, 2, 3], (4, 5, 6), {"key": "value"})

# 27. List of random numbers
import random
lst = random.sample(range(1, 101), 5)
lst.sort()
print(lst)

# 28. Element at index 3
print(["one", "two", "three", "four", "five"][3])

# 29. Combine two dictionaries
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
d1.update(d2)
print(d1)

# 30. Convert list to set
print(set(["apple", "banana", "cherry"]))
```
